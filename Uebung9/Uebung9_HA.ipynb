### IntegratePolynomial

```python
def IntegratePolynomial( p, a, b ):
    result = 0.0
    
    for i in range(len(p)):
        val = p[i] * ( (1/(i+1)) * (b**(i+1) - a**(i+1)))
        result = result + val 
    
    return result
```
