```python
import numpy as np
import matplotlib.pyplot as plt
```

### Explizit
$
\begin{equation}
x(t+\delta t) \approx x(t) + \delta t \: f(x(t),t)
\end{equation}
$
#### Ohne Kühlung
$
\begin{equation}
\rho c \theta (t) = \dot{h_0} exp(\frac{-k}{\theta})
\end{equation}
$


```python
Ntsteps = [1,5,20,100,500]
rho     = 1000  # kg/m³
c       = 1000  # J/kgK
doth    = 1.7e7 # J/sm³
k       = 561   # K
t0      = 0
Tmax    = 100
qab     = 0

fig1 = plt.figure()
ax1 = fig1.gca()
ax1.set_title('Wärmeentwicklung im Reaktor')
ax1.set_xlabel('t [s]')
ax1.set_ylabel(r'$\theta$ [k]')

for N in Ntsteps:
    dt = (Tmax-t0)/N
    T  = np.linspace(t0,Tmax,N+1)
    
    # Theta initialisieren
    theta    = np.zeros(N+1)
    theta[0] = 293  # K 
    
    for j in range(0,N):
         theta[j+1] = theta[j] + dt*((doth*np.exp( -k/theta[j] )/(rho*c)) - qab/(rho*c)) # Iterationsvorschrift
    
    ax1.plot(T, theta,label='{} Zeitschritte'.format(N)) # akutelles Ergebnis in das bestehende Schaubild zeichnen
ax1.legend()
plt.show()
```
